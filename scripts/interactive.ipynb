````python
# read in an example file
import os
import pandas as pd

with open(os.path.join("..","standardHistory","states","9-Czechoslovakia.txt")) as f:
    x = f.readlines()
````

````python
x
````

````python

def getOwnerAndResources(x:list[str]) -> tuple[str, dict]:

    # for each file, work out the owner and resources within
    stateOwner = None
    resources = dict()

    for lineNumber,line in enumerate(x):
        # identify the owner
        line.replace("\t", "")
        loc = line.find("owner = ")
        if loc >=0:
            #found location, get trigraph
            if stateOwner is None:
                stateOwner = line[loc+8:loc+11]
        
        #identify any resources
        resLoc = line.replace(" ","").find("resources={")
        #TODO make the above work also for "resources = {", which is how the custom save editor works
        if resLoc >=0:
            resStr = []
            resLine = line
            resLineNumber = lineNumber

            # found resources. Now we need to read through the next lines until we get to a closed curly brace to identify all the remaining resources.
            while(resLine.find("}") == -1):
                resStr.append(resLine)
                resLineNumber += 1
                if resLineNumber >= len(x):
                    # don't go out of size of x
                    break
                resLine = x[resLineNumber]
            
            #print("resStr:")
            #print(resStr)

            # we now have the resources snippet. Now to parse and add to dictionary
            for subLine in resStr[1:]:
                #split on = and add to dict
                #print("subline:", subLine)
                if subLine.find("=") == -1:
                    #print("= not found, skipping")
                    continue

                resSplit = subLine.split("=")
                res = resSplit[0].lstrip("\t").replace(" ","")

                resSplitQuant = resSplit[1].split()[0]
                #print(f"sl:{subLine},res:{res}, ressplitquant:{resSplitQuant}")
                resSplitQuant = int(float(resSplitQuant))
                #print(resSplitQuant)
                if res not in resources:
                    resources[res] = int(resSplitQuant)
                else:
                    resources[res] = resources[resSplit] + int(resSplitQuant)
                resources.update()
    return [stateOwner, resources]
country, res = getOwnerAndResources(x)

print(country)
print(res)
````

Ok, works on the dummy file. Lets see how it does for all:

````python
HistoricCountryResourceList = dict()
# this is a dictionary with the key as the country trigraph (FRA) and value is a dict of the resources (steel:5,iron:4,...)

for historyFile in os.listdir(os.path.join("..","standardHistory","states")):
    #print(classicHistoryFile)
    with open(os.path.join("..","standardHistory","states",historyFile),"r") as f:
        x = f.readlines()
        country, res = getOwnerAndResources(x)
        #print(country)
        #print(res)
        if country not in HistoricCountryResourceList:
            val = {country:res}
            HistoricCountryResourceList.update(val)
        for resKey,resVal in res.items():
            #print(f"resource:{resKey}")
            if resKey not in HistoricCountryResourceList[country]:
                #print(f"adding {resVal} {resKey} to {country}")
                HistoricCountryResourceList[country].update(res)
            else:
                HistoricCountryResourceList[country][resKey] = HistoricCountryResourceList[country][resKey] + resVal
````

````python
CustomCountryResourceList = dict()
for historyFile in os.listdir(os.path.join("..","history","states")):
    #print(classicHistoryFile)
    with open(os.path.join("..","history","states",historyFile),"r") as f:
        x = f.readlines()
        country, res = getOwnerAndResources(x)
        if country == "ENG":
            print(country)
            print(res)
        if country not in CustomCountryResourceList:
            val = {country:res}
            CustomCountryResourceList.update(val)
        for resKey,resVal in res.items():
            #print(f"resource:{resKey}")
            if resKey not in CustomCountryResourceList[country]:
                #print(f"adding {resVal} {resKey} to {country}")
                CustomCountryResourceList[country].update(res)
            else:
                CustomCountryResourceList[country][resKey] = CustomCountryResourceList[country][resKey] + resVal
````

````python
# let's see how the spread of steel

sSum = 0
historicSteelCount = []
for place,value in HistoricCountryResourceList.items():
    sSum += value.get('steel',0)
    historicSteelCount.append(value.get('steel',0))
print(sSum)

sSum = 0
customSteelCount = []
for place,value in CustomCountryResourceList.items():
    sSum += value.get('steel',0)
    customSteelCount.append(value.get('steel',0))
print(sSum)
````
